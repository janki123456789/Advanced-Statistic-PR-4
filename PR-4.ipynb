### Que 4

A Bernoulli distribution is the simplest probability distribution.

It describes a random experiment with only two possible outcomes:

-  Success (1) with probability p
-  Failure (0) with probability 1-p

Formula:

Probability Mass Function

$$
P(X=x) = p^x (1-p)^{1-x}
$$

Where:

- x = 0 or 1
- p = probability of success


### Que 5
-------

The Binomial distribution is an extension of Bernoulli.

It models the probability of getting k successes in n independent Bernoulli trials.

Formula :
$$
P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}
$$

Where:

- n = total number of trials
- k = number of successes
- p = probability of success
- (ð‘› ð‘˜) = combinations

Combination formula:
$$
\binom{n}{k} = \frac{n!}{k!(n-k)!}
$$

Example:

A customer makes 5 transactions.

- Probability of success transaction p=0.7
- We want probability of 3 successful transactions.

So:

  n=5, 
  k=3,
  p=0.7

$$
P(X=3) = \binom{5}{3} (0.7)^3 (0.3)^2
$$

The Box-Cox Transform is a statistical technique used to make data more normally distributed.

It is commonly applied when data is skewed

It helps in:

- Stabilize variance
- Improve performance of regression models
- Reduce skewness

The transformation only works for positive values.

Formula :

$$
y(\lambda) =
\begin{cases}
\frac{x^\lambda - 1}{\lambda}, & \lambda \neq 0 \\
\ln(x), & \lambda = 0
\end{cases}
$$

Where:

- x = original data
- Î» = transformation parameter
- y(Î») = transformed value
- ln = natural logarithm

```python
import pandas as pd
import numpy as np
from scipy.stats import bernoulli, binom
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Advanced maths/project/PR-4/spread_locator_dataset.csv")
df.head()
```

```python
# Bernoulli Distribution

df['transaction_binary'] = df['transaction_status'].map({'Success':1, 'Fail':0})

p = df['transaction_binary'].mean()

print("\nBernoulli Distribution")
print("Probability of Success (p):", p)

p_success = bernoulli.pmf(1, p)
p_failure = bernoulli.pmf(0, p)

print("P(Success):", p_success)
print("P(Failure):", p_failure)

# Binomial Distribution

df['transaction_date'] = pd.to_datetime(df['transaction_date'])
df['week'] = df['transaction_date'].dt.isocalendar().week

weekly_counts = df.groupby('week')['transaction_id'].count()

print("\nWeekly Transaction Counts")
print(weekly_counts)

n = weekly_counts.max()   
p = df['transaction_binary'].mean()

k = 5
probability = binom.pmf(k, n, p)

print("\nBinomial Distribution")
print(f"Probability of {k} transactions in a week:", probability)

plt.figure()
weekly_counts.plot(kind='bar')
plt.title("Weekly Transaction Count")
plt.xlabel("Week")
plt.ylabel("Number of Transactions")
plt.show()
```

### Que 2

```python
from scipy.stats import poisson

df['transaction_date'] = pd.to_datetime(df['transaction_date'])

daily_transactions = df.groupby('transaction_date')['transaction_id'].count()
lam = daily_transactions.mean()

print("\nAverage transactions per day (Î»):", lam)

k = 5
prob = poisson.pmf(k, lam)

print(f"\nProbability of {k} transactions in a day:", prob)

plt.figure()
plt.plot(daily_transactions, marker="o")
plt.title("Transactions per Day")
plt.xlabel("Date")
plt.ylabel("Number of Transactions")
plt.show()

x = np.arange(0, daily_transactions.max()+1)
poisson_probs = poisson.pmf(x, lam)

plt.figure()
plt.plot(poisson_probs, marker="o")
plt.title("Poisson Distribution Fit")
plt.xlabel("Number of Transactions")
plt.ylabel("Probability")
plt.show()
```

### Que 3

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm
import scipy.stats as stats

amounts = df['transaction_amount']
 
# Log-Normal Distribution

shape, loc, scale = lognorm.fit(amounts, floc=0)

plt.figure()
plt.hist(amounts, bins=30, density=True)

x = np.linspace(min(amounts), max(amounts), 100)
pdf = lognorm.pdf(x, shape, loc, scale)
plt.plot(x, pdf)

plt.title("Log-Normal Fit for Transaction Amount")
plt.xlabel("Transaction Amount")
plt.ylabel("Density")
plt.show()

# Power Law Distribution

counts, bins = np.histogram(amounts, bins=30)
x = (bins[:-1] + bins[1:]) / 2

mask = counts > 0
x = x[mask]
counts = counts[mask]

log_x = np.log(x)
log_y = np.log(counts)

slope, intercept, r_value, p_value, std_err = stats.linregress(log_x, log_y)

plt.figure()
plt.scatter(log_x, log_y)

plt.plot(log_x, intercept + slope*log_x)
plt.title("Power Law")
plt.xlabel("log(Transaction Amount)")
plt.ylabel("log(Frequency)")
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import zscore
from scipy.stats import norm

amount = df["transaction_amount"]

# Calculate Z-Scores

df["Z_score"] = zscore(amount)
print(df[["transaction_amount", "Z_score"]].head())

mean = amount.mean()
std = amount.std()
z_value = (5000 - mean) / std
probability = 1 - norm.cdf(z_value)

print("\nMean:", mean)
print("Standard Deviation:", std)
print("Z-score for 5000:", z_value)
print("Probability:", probability)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

data = df["transaction_amount"]

mean = data.mean()
std = data.std()

x = np.linspace(min(data), max(data), 100)

pdf = norm.pdf(x, mean, std)
cdf = norm.cdf(x, mean, std)

# Plot PDF

plt.figure()
plt.plot(x, pdf)
plt.title("PDF of Transaction Amounts")
plt.xlabel("Transaction Amount")
plt.ylabel("Density")
plt.show()

# Plot CDF

plt.figure()
plt.plot(x, cdf)
plt.title("CDF of Transaction Amounts")
plt.xlabel("Transaction Amount")
plt.ylabel("Cumulative Probability")
plt.show()
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

data = df["transaction_amount"]

norm_params = stats.norm.fit(data)
lognorm_params = stats.lognorm.fit(data)

x = np.linspace(min(data), max(data), 100)

plt.figure()
plt.hist(data, bins=30, density=True)
plt.plot(x, stats.norm.pdf(x, *norm_params), label="Normal Fit")
plt.plot(x, stats.lognorm.pdf(x, *lognorm_params), label="Log-Normal Fit")

plt.legend()
plt.title("Distribution Fit Comparison")
plt.xlabel("Transaction Amount")
plt.ylabel("Density")
plt.show()
```
